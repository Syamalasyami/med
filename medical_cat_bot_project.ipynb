````python
import sys

# Define the function for answering multiple questions
def ask_questions():
    while True:
        user_input = input("Input query: ")

        if user_input.lower() == 'exit':
            print("Exiting...")
            sys.exit()

        if user_input == "":
            continue

        # Check output of the retriever
        retrieved_docs = retriever.get_relevant_documents(user_input)
        print("Retrieved Documents:", retrieved_docs)

        # Check how the prompt is formatted
        formatted_prompt = prompt.format(query=user_input)
        print("Formatted Prompt:", formatted_prompt)

        # Get the response from the LLM
        response = llm.create_chat_completion(messages=[{"role": "user", "content": formatted_prompt}])
        print("LLM Response:", response['choices'][0]['message']['content'])

# Start the Q&A loop
ask_questions()
````
